```python
import folium
import json                                       
import os
import pandas as pd
from sklearn.preprocessing import MinMaxScaler 
from matplotlib import cm
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
f = open('geo_seoul.json', encoding='utf8')
geo_seoul = json.load(f)
f.close()
```

# 데이터 전처리

```python
df = pd.read_csv('project1_df2_3.csv')

df0 = df[df.address.str.contains('서울')==True]
df1 = df0[df0.name.str.contains('Seoul, Korea')==False]
df2 = df1[df1.name.str.contains('Seoul, South Korea')==False]
df3 = df2[df2.address!=(df0[df0.name=='Seoul, Korea'].address.value_counts().keys()[0])]
df4 = df3[df3.time.str.contains('2021')==True]
```

### address에서 구 단위로 추출

```python
address_list = df4.address.reset_index().drop("index", axis=1)["address"]
address_list = list(address_list)

for x in range(len(address_list)):
    print(address_list[x].split()[1])

city = []
for x in range(len(address_list)):
    city.append(address_list[x].split()[1])
df4["city"] = city
```

### 구 단위로 그루핑후  likes 연산

```python
df5 = df4.groupby('city')['likes'].sum()

df5.sort_values(ascending=True)
```

# Folium Choropleth 시각화

```python
map = folium.Map(location=[37.5502, 126.982],
                    tiles='Stamen Toner',
                    zoom_start=11)
bins = [509, 1600, 3000, 5000, 10000, 20000, 37463]
folium.Choropleth(geo_data=geo_seoul,
                  data=df5,
                  columns=('likes'),
                  fill_opacity=0.5,
                  fill_color='PiYG_r',
                  bins=[float(x) for x in bins],
                  key_on='feature.id',
                  legend_name='LIKES',).add_to(map)
map
```
